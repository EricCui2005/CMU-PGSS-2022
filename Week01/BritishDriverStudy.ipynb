```python
import pandas as pd
df = pd.read_csv ("/gdrive/MyDrive/PGSS /Week 1/7 1 2022/britishSeatBeltStudy.csv")


df.columns = ['DriversKilled', 'Drivers', 'Front', 'Rear', 'kms', 'PetrolPrice', 'VanKilled', 'Law', 'Date']

df['Law'] = df['Law'].astype('category')

df['FractionOfDriversKilled'] = df['DriversKilled'] / df['Drivers']

df
```

```python
df.plot.scatter(x = 'Date', y = 'FractionOfDriversKilled', c = 'Law', colormap = 'viridis')
```
